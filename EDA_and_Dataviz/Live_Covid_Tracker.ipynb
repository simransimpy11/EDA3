## About this data

- This data is scraped from the official website of Ministry of Health and Family Welfare (Govt. of India)
- The data gets updated every 24 hours
- It represents number of Covid cases of all the states of India
- The data has been used visualize the trend using matplotlib and seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df.drop("sno", inplace=True, axis=1)
```

```python
df = df[:-1]
```

```python
df
```

```python
plt.figure(figsize=(35,30))
sns.barplot(df['active'], df['state_name'], data=df, order=df.sort_values('active').state_name)
sns.set(font_scale=5)

# set labels
plt.xlabel("Active Cases", size=30)
plt.ylabel("States", size=30)
plt.title("State vs. Active Covid Cases", size=30)
plt.tight_layout()
```

## Top 5 States for active cases

1. Maharashtra
2. Karnataka
3. Andhra Pradesh
4. Uttar Pradesh
5. Kerala

```python
plt.figure(figsize=(35,30))
sns.barplot(df['cured'], df['state_name'], data=df, order=df.sort_values('cured').state_name)
#sns.set(font_scale=2)

# set labels
plt.xlabel("Cured Cases", size=30)
plt.ylabel("States", size=30)
plt.title("State vs. cured Covid Cases", size=30)
plt.tight_layout()
```

## Top 5 States for cured cases

1. Maharashtra
2. Andhra Pradesh
3. Tamil Nadu
4. Karnataka
5. Uttar Pradesh

```python
plt.figure(figsize=(35,30))
sns.barplot(df['death'], df['state_name'], data=df, order=df.sort_values('death').state_name)
#sns.set(font_scale=2)

# set labels
plt.xlabel("Death Cases", size=30)
plt.ylabel("States", size=30)
plt.title("State vs. Death Cases", size=30)
plt.tight_layout()
```

```python
# Reference link : https://python-graph-gallery.com/12-stacked-barplot-with-matplotlib/

plt.figure(figsize=(35,30))

# barplots have to plotted from larger to smaller values
# Otherwise it will overlap and will not be seen
# Here positive > cured > active > death

df_pos = df.sort_values('positive')

barWidth = 1
 
plt.barh(df_pos['state_name'], df_pos['positive'], color='blue', edgecolor='white')
plt.barh(df['state_name'], df['cured'], color='green', edgecolor='white')
plt.barh(df['state_name'], df['active'], color='yellow', edgecolor='white')
plt.barh(df['state_name'], df['death'], color='red', edgecolor='white')
 
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(35,30))

df_newpos = df.sort_values('new_positive')

barWidth = 1
 
plt.barh(df_newpos['state_name'], df_newpos['new_positive'], color='blue', edgecolor='white')
plt.barh(df['state_name'], df['new_cured'], color='green', edgecolor='white')
plt.barh(df['state_name'], df['new_active'], color='yellow', edgecolor='white')
plt.barh(df['state_name'], df['new_death'], color='red', edgecolor='white')
 
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df.corr(), annot=True)
```
